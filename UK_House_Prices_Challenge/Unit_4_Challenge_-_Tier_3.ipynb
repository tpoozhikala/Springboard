**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
#Created a function to take for a Borough name given, the mean average price from a starting year and ending year and providing a price ratio
def create_price_ratio(borough_name,starting_year=2003,end_year=2023):
    """Function to create a price ratio from mean Average Price of starting year and mean Average Price for end year for a London Borough input"""
    london_filtered = london[london['London_Borough'] == str(borough_name)]
    london_start_year = london_filtered[london_filtered['Year'] == starting_year].iloc[:,3]
    london_end_year = london_filtered[london_filtered['Year'] == end_year].iloc[:,3]
    year_ratio = np.mean(london_start_year) / np.mean(london_end_year)
    return year_ratio   
```

```python
#Checking function with Camden London Borough
create_price_ratio('Camden')  
```

```python
#Retreiving column of just the London Boroughs and creating a new dataframe of just the unique London Borough names
boroughs = london['London_Borough'].drop_duplicates()
boroughs_df = pd.DataFrame(boroughs)
boroughs.shape[0]
```

```python
#From the above list, intiated a for loop to calculate price ratios for each London Borough provided and appends to a ratio list
#The list of ratios is then added as a new column in the above boroughs dataframe to create a new dataframe of each London Borough with its associated price ratio
ratio_list = []
for i in boroughs:
    ratio = create_price_ratio(i)
    ratio_list.append(ratio)
boroughs_df['Ratios'] = np.array(ratio_list)
boroughs_df = boroughs_df.dropna()
boroughs_df.head()
```

```python
#Bar plot created to show all the price ratios for each unique London Borough 
boroughs_df.plot(kind='bar', x='London_Borough', y='Ratios')
plt.title('Average Price Ratios from 2003 and 2023 for each London Borough')
plt.ylabel('Average Price Ratios from 2003 and 2023')
plt.show()
```

```python
#Sorting to retrieve the highest values for the ratios
boroughs_df.sort_values('Ratios',ascending=False).head()
```

```python
#Sorting to retrieve the lowest values for the ratios
boroughs_df.sort_values('Ratios',ascending=True).head()
```
