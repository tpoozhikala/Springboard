These are your tasks for this mini project:

1. Collect data from the Franfurt Stock Exchange, for the ticker AFX_X, for the whole year 2017 (keep in mind that the date format is YYYY-MM-DD).
2. Convert the returned JSON object into a Python dictionary.
3. Calculate what the highest and lowest opening prices were for the stock in this period.
4. What was the largest change in any one day (based on High and Low price)?
5. What was the largest change between any two days (based on Closing Price)?
6. What was the average daily trading volume during this year?
7. (Optional) What was the median trading volume during this year. (Note: you may need to implement your own function for calculating the median.)

```python
#Retreiving all ticker price for year 2017 and converting to json object stored as a Python Dict
url = "https://data.nasdaq.com/api/v3/datasets/FSE/AFX_X/data.json?start_date=2017-01-01&end_date=2017-12-31&api_key=VKfkG3Sy56qMP3buWZFD"
res = requests.get(url)
json_data = res.json()
#json_df = pd.DataFrame(json_data)
print(json_data)
```

```python
#Question 3 Solution: Index with data_set data key and 'Open' column in order to open data for the dict object
open_list = []
json_dataset = json_data['dataset_data']['data']

for i in range(0,255):
     open_list.append(json_dataset[i][1])

print(open_list)

#Remove None values in list: open_list
open_list_corrected = []
for val in open_list:
    if val != None :
        open_list_corrected.append(val)

#Calculate highest and lowest prices from the list of open prices
lowest = min(open_list_corrected)
highest = max(open_list_corrected)

print(f"The highest opening price is: ${highest} and the lowest opening price is: ${lowest}")
```

```python
#Question 4 Solution: Index with data_set data key and 'Change' column in order to open data for the dict object
change_list = []
json_dataset = json_data['dataset_data']['data']

for i in range(0,255):
     change_list.append(json_dataset[i][3] - json_dataset[i][4])

print(change_list)

#Remove None values in list: change_list
change_list_corrected = []
for val in change_list:
    if val != None :
        change_list_corrected.append(val)

#Calculate largest change in 'change' column
negative_change = min(change_list_corrected)

print(f"The largest change in any one day from Highest to Lowest price is: ${negative_change}")
```

```python
#Question 5 Solution: Index with data_set data key and 'Change' column in order to open data for the dict object
change_2_days_list = []
json_dataset = json_data['dataset_data']['data']

for i in range(0,255):
    for j in range(1,255):
         change_2_days_list.append(json_dataset[i][4] - json_dataset[j][4])

print(change_2_days_list)

#Remove None values in list: change_list
#change_list_corrected = []
#for val in change_list:
#    if val != None :
#        change_list_corrected.append(val)

#Calculate largest change in 'change' column
max_change = max(change_2_days_list)

print(f"The largest change between any 2 days based on closing price is: ${max_change}")
```

```python
#Question 6 Solution: Index with data_set data key and 'Traded_vol' column in order to open data for the dict object:
traded_list = []
json_dataset = json_data['dataset_data']['data']

for i in range(0,255):
     traded_list.append(json_dataset[i][6])

print(traded_list)

#Calculate average of the traded list column:
average_traded_volume = sum(traded_list) / len(traded_list)

print(f"The average traded volume during 2017 is: ${average_traded_volume}")
```

```python
#Question 7 Solution: Index with data_set data key and 'Traded_vol' column in order to open data for the dict object:
traded_list = []
json_dataset = json_data['dataset_data']['data']

for i in range(0,255):
     traded_list.append(json_dataset[i][6])

print(traded_list)

#Calculate average of the traded list column:
median_traded_volume = np.median(traded_list)

print(f"The median traded volume during 2017 is: ${median_traded_volume}")
```
