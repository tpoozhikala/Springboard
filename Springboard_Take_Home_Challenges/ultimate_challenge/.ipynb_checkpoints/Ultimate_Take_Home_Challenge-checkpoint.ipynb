### 1.2 Load/Transform Dataset

Began to explore the dataset by importing the json file and then converting the json datetime entry with .fromisoformat to form into entries readable to be transformed into a pandas dataframe.

```python
#Code found to read json data into json.loads on https://www.freecodecamp.org/news/loading-a-json-file-in-python-how-to-read-and-parse-json/
path = 'C:/Users/tpooz/OneDrive/Desktop/Data_Science_BootCamp_2023/SpringBoard_Github/Springboard/Springboard_Take_Home_Challenges/ultimate_challenge/logins.json'
with open(path) as file:
    file_content = file.read()
date = json.loads(file_content)
```

```python
#Code to convert each json entry datetime into Python datetime object found on https://pythonkb.com/python-convert-json-date-string-to-python-datetime/
login_list = []
for i in range(len(date['login_time'])):
    login_list.append(datetime.fromisoformat(date['login_time'][i]))
login_list
```

```python
login_df= pd.DataFrame(login_list, columns=['login_time'])
login_df
```

Checked the dataset if any 'null' values appeared within the dataset.

```python
login_df.isnull().sum()
```

In order to format the dataset to be further aggregated of count of logins within 15 min intervals, created a new field labeled 'Count' and store each entries value as a value of 1.

```python
login_df['Count'] = 1
login_df
```

Since the 'login_time' field is set as datetime fields, I set these values as the index in order to begin aggregating the dataset to get counts of logins within 15 min intervals.

```python
#Code found to set index of dataframe on https://www.statology.org/pandas-group-by-5-minute-intervals/
login_df = login_df.set_index('login_time')
login_df
```

```python
#Code found to aggregate datetime data within specific interval found on https://www.statology.org/pandas-group-by-5-minute-intervals/
agg_df = login_df.resample('15min').sum()
agg_df.head()
```

Now the dataset is aggregated for every 15 min interval and ready for further analysis.

## 2. Experiment and Metrics Design

### 2.1 Introduction

From given challenge prompt:

The neighboring cities of Gotham and Metropolis have complementary circadian rhythms: on weekdays, Ultimate Gotham is most active at night, and Ultimate Metropolis is most active during the day. On weekends, there is reasonable activity in both cities. However, a toll bridge, with a two way toll, between the two cities causes driver partners to tend to be exclusive to each city. The Ultimate managers of city operations for the two cities have proposed an experiment to encourage driver partners to be available in both cities, by reimbursing all toll costs.

### 2.2 Question 1:

1) What would you choose as the key measure of success of this experiment in
encouraging driver partners to serve both cities, and why would you choose this metric?

A key metric of success that I would choose for this experiment in terms of encouraging driver partners to serve both cities would assessing total number of cars crossing the toll bridge between both cities. If the number of cars crossing the toll bridge after applying the above experiment increases by a statiscally signifcant amount, then we can reject the supposed null hypothesis that removing the toll is not statiscally signifcant. However, further analysis will need to be conducted to determine if in fact removing the toll is the main cause, but if the number of cars increases by a statiscally signifcant amount, then at least this metric provides an understanding as per next steps.

### 2.3 Question 2:

2) Describe a practical experiment you would design to compare the effectiveness of the
proposed change in relation to the key measure of success. Please provide details on:
a) how you will implement the experiment
b) what statistical test(s) you will conduct to verify the significance of the
observation
c) how you would interpret the results and provide recommendations to the city
operations team along with any caveats.

Information regarding statiscal testing found on: https://www.scribbr.com/statistics/statistical-tests/

The practical experiment that I would design in order to assess number of cars crossing the toll bridge would be to implement a car counter sensor above the bridge connecting each of the cities. The counter's data would be stored within either cities Datawarehouse and ownership of the Datawarehouse would be determined within the legal contract between the each city's partnership. The statistical test to be performed would be an independent t-test with the null hypothesis being that the average car count for group of cars crossing the bridge in the previous year equals the average car count after the year of implementation set with an alpha value of 0.05. This means that if the difference between the means of the group of number of cars from the previous year and the year after reimbursing the toll costs has a p-value less than 0.05, then we can reject the null hypothesis and say that there in fact is a statistically significant difference between each of the groups. Based on these results, this would help the operations team understand if removing the toll costs increased overall driver partnerships between each of the cities. Though further analysis will need to be undertaken as the following tests only provides whether or not there is a statistically significant difference between the groups, not whether which group was higher or lower and this experiment only provides statistical significance not overall conclusive evidence of behavior. Possibly data would need to be sampled for longer than a year to have further confidence in the concretely implementing the strategy of reimbursing toll costs.

### 3.2 Load/Transform Dataset

Loading dataset by first converting json data into pandas dataframe.

```python
#Code found to read json data into json.loads on https://www.freecodecamp.org/news/loading-a-json-file-in-python-how-to-read-and-parse-json/
path = 'C:/Users/tpooz/OneDrive/Desktop/Data_Science_BootCamp_2023/SpringBoard_Github/Springboard/Springboard_Take_Home_Challenges/ultimate_challenge/ultimate_data_challenge.json'
with open(path) as file:
    file_content = file.read()
data = json.loads(file_content)
```

```python
retention_df = pd.DataFrame(data)
retention_df
```

Replacing column 'ultimate_black_user' information from True and False to 1 or 0 for future modeling steps.

```python
retention_df['ultimate_black_user'] = retention_df['ultimate_black_user'].replace({True:1, False:0})
retention_df
```

Created column 'active' to capture values of users who took a trip within the first 30 days of signing up as mentioned by the prompt.

```python
#Code to create a filter and replace True and False value with 0 and 1 found on https://stackoverflow.com/questions/26830752/making-new-column-in-pandas-dataframe-based-on-filter
retention_df['active'] = retention_df['trips_in_first_30_days'] >= 1
```

```python
retention_df['active'] = retention_df['active'].replace({True:1, False:0})
```

```python
retention_df
```

Converting the 'signup_date' and 'last_trip_date' columns into datetime fields and creating a new column 'date_diff' taking the difference between these two dates as an added feature for model development.

```python
sign_up = pd.to_datetime(retention_df['signup_date'])
retention_df['signup_date'] = sign_up
last_trip = pd.to_datetime(retention_df['last_trip_date'])
retention_df['last_trip_date'] = last_trip
```

```python
retention_df['date_diff'] = retention_df['last_trip_date'] - retention_df['signup_date']
retention_df
```

```python
sign_up = pd.to_datetime(retention_df['signup_date'])
retention_df['signup_date'] = sign_up
last_trip = pd.to_datetime(retention_df['last_trip_date'])
retention_df['last_trip_date'] = last_trip
```

```python
#Code to convert pandas dataframe column to string found on https://stackoverflow.com/questions/22005911/convert-columns-to-string-in-pandas
retention_df['date_diff'] = retention_df['last_trip_date'] - retention_df['signup_date']
retention_df['date_diff'] = retention_df['date_diff'].astype(str).str.strip('days').astype(float)
retention_df
```

Created another feature labeled 'active_5_mon' to address prompt ask of developing a model to predict if a user will still be retained after 6 months. I chose 5 months in order to still have reasonable data to feed for model prediction but am making the assumption that if the users stayed on the system after 5 months, they would still be predicted to be retained at 6 months as well.

```python
retention_df['active_5_mon'] = retention_df['date_diff'] > 150
retention_df['active_5_mon'] = retention_df['active_5_mon'].replace({True:1, False:0})
retention_df
```

### 3.5 Predictive Modeling

#### 3.5.1 One-Hot Encoding/Train_Test_Split

Prior to fitting different predictive models, the dataset needs to be one-hot encoded as typically(depending on the model) predictive models need features with all numeric values in the rows.

```python
#Dropping datetime fields and 'active' since creating the model to only predict if a user would be active within 5 months after signup date.
X = retention_df.drop(['active', 'active_5_mon', 'signup_date','last_trip_date'], axis=1)
y = retention_df['active_5_mon']
```

```python
#Avoid dummy variable trap of intercorrleation between one-hot encoded variable by removing the more overrepresented columns of Winterfell and iPhone users.
X = pd.get_dummies(X)
X = X.drop(['city_Winterfell', 'phone_iPhone'], axis=1)
X
```

```python
#Create a train/test split datasets with a 70/30 train/test for training and assessing different predictive models.
X_train, x_test, y_train, y_test = train_test_split(X , y, test_size=0.3, random_state=42)
```

#### 3.5.2 Logistic Regression

Training Logistic Regression Model as well as applying GridSearchCV in order to test different hyperparameters for penalty and perform 5-fold cross validation to provide a more generalizable model.

```python
logreg = LogisticRegression(max_iter=350)
params_logreg = {'penalty':['l1','l2','elasticnet', 'none'], 'solver':['lbfgs','newton-cg','saga'], 'C':[0.001, 0.01, 0.1, 1]}
#params_logreg = {'penalty':['l1','l2','elasticnet', None]}
logreg_grided = GridSearchCV(estimator=logreg, param_grid=params_logreg, scoring='accuracy', cv=5)
```

```python
#Code to record time of code execution found on https://www.geeksforgeeks.org/python-measure-time-taken-by-program-to-execute/
start_log = time.time()
logreg_grided.fit(X_train, y_train)
end_log = time.time()
total_log = end_log - start_log
```

```python
print(f"Logistic Regression fit time: {total_log:.2f} seconds")
```

```python
logreg_grided.best_estimator_.feature_names_in_
```

```python
logreg_grided.best_estimator_.feature_importances_
```

```python
logreg_grided.best_params_
```

Once model is fit, then applying the fit model to predict target variable of 'active_5_mon' with x_test set.

```python
predict_log = logreg_grided.predict(x_test)
```

```python
logreg_grided.best_score_
```

```python
print(confusion_matrix(predict_log, y_test))
```

```python
fpr_log, tpr_log, thresholds_log = roc_curve(predict_log, y_test)
```

```python
logreg_roc = roc_auc_score(predict_log, y_test)
logreg_roc
```

```python
#Code to plot ROC Curve found on https://www.geeksforgeeks.org/auc-roc-curve/
plt.plot(fpr_log, tpr_log, '-o',label='ROC')
plt.plot([0,1],[0,1],'--',label='Baseline')
plt.xlabel('False Postive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve Logistic Regression')
plt.legend()
plt.show()
```
